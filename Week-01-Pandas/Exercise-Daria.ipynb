## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
# 1. How many listings are there with a price less than 100? 
condition= df['price'] < 100
len(df[condition])
```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.

condition1= df['neighbourhood_group']=="Brooklyn"
df_bk= len(df[condition1])
print(df_bk)
```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.
len(df[condition&condition1])

```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
# How many total are there that have those names
names= ["Michael","David","John","Daniel" ]
new_condition=df['host_name'].isin(names)

df[new_condition]['host_name'].value_counts()
```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 

df["adjusted_price"] = np.where(df["neighbourhood"] == "Williamsburg", df["price"] + 100, df["price"])

df
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.  

private=df['room_type']=="Private room"
shared=df['room_type']=="Shared room"

private_amount= len(df[private])
shared_amount=len(df[shared])

total=len(df['room_type'])

percent_private= private_amount/total *100
percent_shared= shared_amount/total *100

print(round(percent_private,2))
print(round(percent_shared,2))
```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.
neighboorhood=df.groupby('neighbourhood_group')
neighboorhood.count()
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 
neighboorhood=df.groupby('neighbourhood_group')
round(neighboorhood['price'].mean(),2)

```

```python
# 2.5. Using `groupby`, find the mean price for each room_type. 
neighboorhood=df.groupby('room_type')
round(neighboorhood['price'].mean(),2)
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 
neighboorhood=df.groupby('neighbourhood_group')
neighboorhood['price'].agg(['max','min'])

```

```python
# 4. Using `groupby`, find the mean price for each room_type in each neighbourhood_group.
neighboorhood=df.groupby(['neighbourhood_group','room_type'])
round(neighboorhood['price'].mean(),2)
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.
neighboorhood=df.groupby(['neighbourhood_group','room_type'])
neighboorhood['price'].agg(['max','min','mean','median','std'])


```

1. Who was won Album of the Year in 2016?


2. Who won Best Rap Album in 2009?

```python
df = pd.read_csv('data/grammys.csv')
df
```

```python
# 1
condition=df['category'] == 'Album of the Year'
condition1= df['year'] == 2016
condition2= df['winner'] == True
album_of_the_year_2016 = df[condition & condition1& condition2]
album_winner_2016 = album_of_the_year_2016['workers'].values[0] 

album_winner_2016
```

```python
# 2
condition=df['category'] == 'Best Rap Album'
condition1= df['year'] == 2009
condition2= df['winner'] == True
winner = df[condition & condition1& condition2]
rap_album_winner_2009= winner['workers'].values[0] 

rap_album_winner_2009
```

```python
# 3
amount=df['workers'] == 'Kendrick Lamar'
amount_nominated=amount.sum()
print("He was nominated", amount_nominated, "times")

condition = (df['workers'] == 'Kendrick Lamar') & (df['winner'] == True)  #i realized just now I could do this haha
final_wins = condition.sum()

print("He actually won", final_wins, "times")
```
